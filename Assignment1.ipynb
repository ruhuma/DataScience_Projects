**First Visualisation**

**Histogram**

```python
classification1 = data[data.Classification == 'Iris-setosa']
classification2 = data[data.Classification == 'Iris-versicolor']
classification3 = data[data.Classification == 'Iris-virginica']
plt.figure()
plt.hist(classification1['PL'],color='red',edgecolor='black',alpha=.9)
plt.hist(classification2['PL'],color='green',edgecolor='yellow',alpha=.7)
plt.hist(classification3['PL'],color='blue',edgecolor='cyan',alpha=.5)
plt.xlabel('Petal Length')
plt.legend(["Iris-setosa","Petal-versicolor","Iris-virginica"])
plt.title('Distribution of Petal Length in 3 species')
```

**Second Visualisation**

**Kernel Density Estimation**

```python
import seaborn as sns

sns.kdeplot(classification1['SL'],fill =True)
sns.kdeplot(classification2['SL'],fill =True)
sns.kdeplot(classification3['SL'],fill =True)
plt.legend(["Iris-setosa","Petal-versicolor","Iris-virginica"])
plt.xlabel('Sepal Length')
plt.title('Distribution of Sepal Length in 3 species')
```

**Third Visualisation**

**Scatter Plots**

```python
x=classification1['PL']
y=classification1['PW']
plt.scatter(x,y,c='Red')
x=classification2['PL']
y=classification2['PW']
plt.scatter(x,y,c='Green')
x=classification3['PL']
y=classification3['PW']
plt.scatter(x,y,c='Blue')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(["Iris-setosa","Iris-versicolor","Iris-virginica"])
plt.title('Petal length vs Petal Width for 3 species')
```

```python
x=classification1['PL']
y=classification1['SL']
plt.scatter(x,y,c='cyan')
x=classification2['PL']
y=classification2['SL']
plt.scatter(x,y,c='red')
x=classification3['PL']
y=classification3['SL']
plt.scatter(x,y,c='yellow')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.legend(["Iris-setosa","Iris-versicolor","Iris-virginica"])
plt.title('Petal length vs Sepal length for 3 species')
```
